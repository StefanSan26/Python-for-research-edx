## Reading the Data

```python
import scipy.stats as ss
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_time = pd.read_csv("train_time_series.csv", index_col=0)
train_time
```

```python
train_labels = pd.read_csv("train_labels.csv", index_col=0)
train_labels
```

## Manipulating The Data

```python
copy_train_labels=pd.DataFrame()
for i in range(len(train_labels)):
	ind=train_labels.iloc[i].name
	copy_train_labels = copy_train_labels.append(train_time.loc[ind])
copy_train_labels
```

```python
copy_train_labels["label"] = train_labels["label"]
copy_train_labels
```

## Model

```python
X_train=copy_train_labels[["x","y","z"]]
X_train
```

```python
Y_train=copy_train_labels["label"]
Y_train
```

```python
forest_classifier = RandomForestClassifier(random_state=1)
forest_classifier.fit(X_train,Y_train)
```

## Test

```python
test_time = pd.read_csv("test_time_series.csv", index_col=0)
test_time
```

```python
test_labels = pd.read_csv("test_labels.csv", index_col=0)
test_labels
```

```python
y_eval=forest_classifier.predict_proba(X_train)
y_eval
```

```python
lbls_eval = []
for i in y_eval:
	lbls_eval.append(np.argmax(i)+1)
lbls_eval
```

```python
(lbls_eval == Y_train).mean()
```

```python

copy_test_time=pd.DataFrame()
for i in range(len(test_labels)):
	ind=test_labels.iloc[i].name
	copy_test_time = copy_test_time.append(test_time.loc[ind])
copy_test_time
```

```python
X_test=copy_test_time[["x","y","z"]]
y_predict=forest_classifier.predict_proba(X_test)
y_predict
```
